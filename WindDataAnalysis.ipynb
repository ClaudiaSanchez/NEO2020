### Dirección del viento 

Podríamos decir que los ángulos están en coordenadas polares (0 - 360)

Para transformar coordenadas polares a rectángulares:
$$ x = r \cos( \theta ) $$
$$ y = r \sin( \theta ) $$

Para transformar de coordenadas rectángulares a polares:
$$ r = \sqrt{ x^2 + y^2 } $$
$$ \theta = \arctan( y, x ) $$

Convertir radianes a grados
$$ \pi radianes = 180 grados $$

```python
fsin = interp1d( idx_167[direccion_167!=0], np.sin(direccion_167[direccion_167!=0]*3.1416/180))
fcos = interp1d( idx_167[direccion_167!=0], np.cos(direccion_167[direccion_167!=0]*3.1416/180))
dir_167_predict_sin = fsin( idx_167 ) 
dir_167_predict_cos = fcos( idx_167 )
dir_167_predict = np.arctan2(dir_167_predict_sin,dir_167_predict_cos)*180/3.1416
dir_167_predict[ dir_167_predict<0 ] = dir_167_predict[ dir_167_predict<0 ]+360

print(direccion_167[:10])
print(dir_167_predict[:10])

plt.figure()
plt.subplot(2,1,1)
plt.title('Dirección', fontsize=8)
plt.plot(idx_167,direccion_167,'.',color='gray')
plt.plot(idx_167[direccion_167==0.0],direccion_167[direccion_167==0.0],'.',color='red')
plt.xticks([])
plt.subplot(2,1,2)
plt.title('Dirección', fontsize=8)
plt.plot(idx_167,dir_167_predict,'.',color='gray')
plt.plot(idx_167[direccion_167==0.0],dir_167_predict[direccion_167==0.0],'.',color='blue')
plt.xticks([])
plt.show()
```
